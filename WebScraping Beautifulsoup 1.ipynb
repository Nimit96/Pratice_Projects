#### 1. Write a python program to display all the header tags from ‘en.wikipedia.org/wiki/Main_Page’.

```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
## url = https://en.wikipedia.org/wiki/Main_Page
```

```python
head=[]
def get_heading(url):
    page= requests.get(url)
    soup = BeautifulSoup(page.content)
    heading= soup.find_all('span', class_='mw-headline')
    for i in heading:
        head.append(i.text.replace('\xa0'," "))
    return head
```

```python
url= input(print("Enter URL \n"))
get_heading(url)
```

#### 2. Write a python program to display IMDB’s Top rated 100 movies’ data (i.e. Name, IMDB rating, Year of release) and make data frame.

```python
## url= https://www.imdb.com/chart/top/
```

```python
def top_movies(url):
    page= requests.get(url)
    soup= BeautifulSoup(page.content)
    movie_name= soup.find_all("td", class_="titleColumn")
    movie=[]
    for i in movie_name:
        movie.append(i.text.replace("\n"," "))
    movie_names= movie[:100]
    movie=[]
    for i in range(100):
        s= movie_names[i]
        s= s[11:]
        s= s[:-7]
        s= s.strip()
        movie.append(s)
    year= soup.find_all('span', class_='secondaryInfo')
    year_=[]
    for i in year:
        year_.append(i.text)
    year_release= year_[:100]
    rate= soup.find_all('td',class_='ratingColumn imdbRating')
    rating=[]
    for i in rate:
        rating.append(i.text.replace('\n',' '))
    imdb_rating=rating[:100]
    Top_movies= pd.DataFrame()
    Top_movies["Movie_Name"]= movie
    Top_movies["Year_of_release"]=year_release
    Top_movies["IMDB_Rating"]= imdb_rating
    return Top_movies
```

```python
url= input(print("Enter URL \n"))
top_movies(url)
```

#### 4. Write a python program to scrap book name, author name, genre and book review of any 5 books from
‘www.bookpage.com’

```python
## url= https://bookpage.com/reviews
```

```python
def book(url):
    page= requests.get(url)
    soup= BeautifulSoup(page.content)
    book=soup.find_all('h4', class_="italic")
    book_name=[]
    for i in book:
        book_name.append(i.text.replace("\n"," "))
    book_name= book_name[:5]
    author=soup.find_all("p", class_="sans bold")
    author_name=[]
    for i in author:
        author_name.append(i.text.replace("\n"," "))
    author_name= author_name[:5]
    genre_= soup.find_all("p", class_="genre-links hidden-phone")
    genre=[]
    for i in genre_:
        genre.append(i.text.replace("\n"," "))
    genre= genre[:5]
    r= soup.find_all("p", class_="excerpt")
    review=[]
    for i in r:
        review.append(i.text.replace("\n"," "))
    review= review[:5]
    Book_reviews= pd.DataFrame()
    Book_reviews["Book_title"]= book_name
    Book_reviews["Author"]= author_name
    Book_reviews["Genre"]= genre
    Book_reviews["Review"]= review
    return Book_reviews
```

```python
url= input("Enter URL \n")
print("Review of any 5 books: \n")
book(url)
```

#### 5. Write a python program to scrape cricket rankings from ‘www.icc-cricket.com’. You have to scrape:

#### i) Top 10 ODI teams in men’s cricket along with the records for matches, points and rating.

```python
## url= https://www.icc-cricket.com/rankings/mens/team-rankings/odi
```

```python
def ODI_teams(url):
    page= requests.get(url)
    soup=BeautifulSoup(page.content)
    team= soup.find_all("span", class_="u-hide-phablet")
    teams=[]
    for i in team:
        teams.append(i.text)
    teams=teams[:10]
    match1=soup.find("td", class_="rankings-block__banner--matches")
    match1=match1.text
    point1=soup.find("td", class_="rankings-block__banner--points")
    point1= point1.text
    rating1=soup.find("td",class_="rankings-block__banner--rating u-text-right")  
    rating1= rating1.text.replace("\n"," ")
    rating1= rating1.strip()
    match_point= soup.find_all("td",class_="table-body__cell u-center-text")
    matches=[]
    for i in match_point:
        matches.append(i.text)
    match=[]
    points=[]
    for i in range(len(matches)):
        if i%2==0:
            match.append(matches[i])
        else:
            points.append(matches[i])
    match.insert(0,match1)
    points.insert(0,point1)
    rating= soup.find_all("td", class_="table-body__cell u-text-right rating")
    ratings=[]
    for i in rating:
        ratings.append(i.text)
    ratings.insert(0,rating1)
    ODI_teams= pd.DataFrame()
    ODI_teams["Team"]=teams
    ODI_teams["Matches"]= match[:10]
    ODI_teams["Points"]=points[:10]
    ODI_teams["Ratings"]=ratings[:10]
    return ODI_teams
```

```python
url= input("Enter URL \n")
print("Top 10 ODI teams in Men's cricket are: \n")
ODI_teams(url)
```

#### ii) Top 10 ODI Batsmen in men along with the records of their team and rating.

```python
## url= https://www.icc-cricket.com/rankings/mens/player-rankings/odi/batting
```

```python
def ODI_player(url):
    page= requests.get(url)
    soup= BeautifulSoup(page.content)
    name= soup.find("div", class_="rankings-block__banner--name-large")
    name=name.text
    names= soup.find_all("td", class_="table-body__cell rankings-table__name name")
    players=[]
    for i in names:
        players.append(i.text.replace("\n"," "))
    players.insert(0,name)
    team1= soup.find("div",class_="rankings-block__banner--nationality")
    team1=team1.text.replace("\n"," ")
    team= soup.find_all("span", class_="table-body__logo-text")
    teams=[]
    for i in team:
        teams.append(i.text.replace("\n"," "))
    teams.insert(0,team1)
    rate1= soup.find("div", class_="rankings-block__banner--rating")
    rate1= rate1.text.replace("\n"," ")
    rating= soup.find_all("td", class_="table-body__cell rating")
    ratings=[]
    for i in rating:
        ratings.append(i.text.replace("\n"," "))
    ratings.insert(0,rate1)
    record1= soup.find("span",class_="rankings-block__career-best-text")
    record1= record1.text.replace("\n"," ")
    record= soup.find_all("td",class_="table-body__cell u-text-right u-hide-phablet")
    records=[]
    for i in record:
        records.append(i.text.replace("\n"," "))
    records.insert(0,record1)
    record=[]
    for i in range(10):
        s= records[i]
        s=s.strip()
        record.append(s)
    ODI_Batsmen= pd.DataFrame()
    ODI_Batsmen["Name"]=players[:10]
    ODI_Batsmen["Team"]= teams[:10]
    ODI_Batsmen["Ratings"]=ratings[:10]
    ODI_Batsmen["Record"]=record[:10]
    return ODI_Batsmen
```

```python
url= input("Enter URL \n")
print("Top 10 ODI batsmen in Men's cricket ---- \n")
ODI_player(url)
```

#### 7. Write a python program to scrape details of all the mobile phones under Rs. 20,000 listed on Amazon.in. The scraped data should include Product Name, Price, Image URL and Average Rating.

```python
def phones(url):
    page= requests.get(url)
    soup= BeautifulSoup(page.content)
    name= soup.find_all("span", class_="a-size-medium a-color-base a-text-normal")
    product_name=[]
    for i in name:
        product_name.append(i.text)
    print(product_name)
    price= soup.find_all("span", class_="a-price-whole")
    price_=[]
    for i in price:
        price_.append(i.text)
    print(price_)
    image= soup.find_all("img", class_="s-image")
    image_url=[]
    for i in image:
        image_url.append(i.get('src'))
    rating= soup.find_all("div",class_="a-row a-size-small")
    ratings=[]
    for i in rating:
        ratings.append(i.text)
    MobilePhone= pd.DataFrame()
    MobilePhone["Phone_Name"]=product_name[:len(price_)]
    MobilePhone["Price"]=price_[:len(price_)]
    MobilePhone["Image URL"]=image_url[:len(price_)]
    MobilePhone["Ratings"]=ratings[:len(price_)]
    return MobilePhone
```

```python
url= input("Enter URL \n")
print("Mobile Phones under 20,000 listed in amazon.in :\n")
phones(url)
```

#### 8. Write a python program to extract information about the local weather from the National Weather Service website of USA, https://www.weather.gov/ for the city, San Francisco. You need to extract data about 7 day extended forecast display for the city. The data should include period, short description, temperature and description.

```python
## url= https://forecast.weather.gov/MapClick.php?lat=37.777120000000025&lon=-122.41963999999996#.YMsaYqgzbIU
```

```python
def weather_info(url):
    page=requests.get(url)
    soup= BeautifulSoup(page.content)
    period_=soup.find_all("div", class_="col-sm-2 forecast-label")
    period=[]
    for i in period_:
        period.append(i.text)
    description=soup.find_all("div", class_="col-sm-10 forecast-text")
    desc=[]
    for i in description:
        desc.append(i.text)
    weather= pd.DataFrame()
    weather["Period"]=period
    weather["Forecast"]=desc
    print("7-day extended forecast for San Francisco is \n")
    return weather
```

```python
url= input(print("Enter url \n"))
weather_info(url)
```

#### 9. Write a python program to scrape fresher job listings from ‘https://internshala.com/’. It should include job title, company name, CTC, and apply date.

```python
## url= https://internshala.com/fresher-jobs
```

```python
def jobs_list(url):
    page= requests.get(url)
    soup= BeautifulSoup(page.content)
    title= soup.find_all("div",class_="heading_4_5 profile")
    job_title=[]
    for i in title:
        job_title.append(i.text.replace("\n"," "))
    comp= soup.find_all("div",class_="heading_6 company_name")
    comp_name=[]
    for i in comp:
        comp_name.append(i.text.replace("\n"," "))
    for i in range(len(comp_name)):
        comp_name[i]= comp_name[i].strip()
    ctc_= soup.find_all("div", class_="item_body")
    ctc=[]
    for i in ctc_:
        ctc.append(i.text.replace("\n"," "))
    list=[]
    for i in range(len(ctc)):
        if i%3==0:
            continue
        else:
            list.append(ctc[i])
    ctc=[]
    apply=[]
    for i in range(len(list)):
        if i%2==0:
            ctc.append(list[i].strip())
        else:
            apply.append(list[i])
    ## Constructing a DataFrame
    print("Jobs for freshers listed in internshala.com are :\n")
    jobs= pd.DataFrame()
    jobs["Job_Title"]=job_title
    jobs["Company_Name"]=comp_name
    jobs["CTC"]=ctc
    jobs["Apply By"]=apply
    return jobs
```

```python
url= input(print("Enter URL \n "))
jobs_list(url)
```
